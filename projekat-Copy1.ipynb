#### Statistika podataka

```python
def normalizacija_mag(mag):
    return (mag - mag.min()) / (mag.max() - mag.min())

def amplituda(ime, mag):
    for i in range(len(ime)):
        ime_tmp = ime[i]
        mag_tmp = mag[i]
        amplituda = np.max(mag_tmp) - np.min(mag_tmp)
        print(f"{ime[i]} – amplituda: {amplituda:.3f}")
        #print("-" * 40)

def varijansa(ime,mag):
    for i in range(len(ime)):
        ime_tmp = ime[i]
        mag_tmp = mag[i]
        varijansa = np.var(mag_tmp)
        print(f"{ime[i]} – varijansa: {varijansa:.3f}")
        #print("-" * 40)

#iskosenost i zasiljenost
def skewness(ime,mag):
    for i in range(len(ime)):
        ime_tmp = ime[i]
        mag_tmp = mag[i]
        iskosenost = skew(mag_tmp)
        print(f"{ime[i]} – iskosenost: {iskosenost:.3f}")
        #print("-" * 40)

def kurt(ime,mag):
    for i in range(len(ime)):
        ime_tmp = ime[i]
        mag_tmp = mag[i]
        zasiljenost = kurtosis(mag_tmp)
        print(f"{ime[i]} – zasiljenost: {zasiljenost:.3f}")
        #print("-" * 40)
```

#### vizuelizacija 

```python
def dominantni_periodi(ime, vreme, mag, prominence=0.005):
    n = len(ime)
    nrows, ncols = 3, 4
    fig, ax = plt.subplots(nrows, ncols, figsize=(20, 12))
    ax = ax.flatten()

    ls_periodi = []
    ls_snage = []
    
    for i in range(n):
        vreme_tmp = vreme[i]
        mag_tmp = mag[i]

        frequency, power = LombScargle(vreme_tmp, mag_tmp).autopower()
        periods = 1 / frequency

        peaks, _ = find_peaks(power, prominence=prominence)
        dominant_periods = periods[peaks]
        dominant = periods[np.argmax(power)]
        dominant_power = power[np.argmax(power)]
        
        ls_periodi.append(dominant)
        ls_snage.append(dominant_power)
        
        # Plot u subplot
        ax[i].plot(periods, power, color='indigo', label='Periodogram')
        ax[i].axvline(dominant, color='crimson', linestyle='--', label=f'Dominantni ≈ {dominant:.2f} d')

        for p in dominant_periods[:3]:
            ax[i].axvline(p, color='orange', linestyle=':', alpha=0.5)

        ax[i].set_title(f"{ime[i]}")
        ax[i].set_xlabel("Period (dani)")
        ax[i].set_ylabel("Snaga")
        #ax[i].set_xlim(800,1200)
        ax[i].legend(fontsize=8)
        ax[i].grid(True)

        print(f"{ime[i]} – Dominantni period: {dominant:.3f} dana")
        print(f"{ime[i]} – snaga perioda: {dominant_power:.3f}")
        print("Top periodi (dani):", np.round(dominant_periods[:3], 3))
        print("-" * 40)

    for j in range(n, len(ax)):
        ax[j].set_visible(False)

    plt.tight_layout()
    plt.show()
    return ls_periodi,ls_snage

_,_ = dominantni_periodi(agn_ime, agn_vreme, agn_norm_mag)
```

#### T-test,KS2,MannWhitneyu

```python
def amplituda(ime, mag):
    amplitude = []
    for i in range(len(ime)):
        mag_tmp = np.array(mag[i], dtype=float)
        if len(mag_tmp) > 0 and not np.isnan(mag_tmp).all():
            amp = np.nanmax(mag_tmp) - np.nanmin(mag_tmp)
            amplitude.append(amp)
        else:
            amplitude.append(np.nan)
    return np.array(amplitude)

def varijansa(ime, mag):
    varijanse = []
    for i in range(len(ime)):
        mag_tmp = np.array(mag[i], dtype=float)
        if len(mag_tmp) > 1 and not np.isnan(mag_tmp).all():
            varij = np.nanvar(mag_tmp)
            varijanse.append(varij)
        else:
            varijanse.append(np.nan)
    return np.array(varijanse)

def skewness(ime,mag):
    skewness = []
    for i in range(len(ime)):
        mag_tmp = np.array(mag[i], dtype=float)
        if len(mag_tmp) > 1 and not np.isnan(mag_tmp).all():
            skew = np.nanvar(mag_tmp)
            skewness.append(skew)
        else:
            skewness.append(np.nan)
    return np.array(skewness)

def kurtosis(ime,mag):
    kurtosis = []
    for i in range(len(ime)):
        mag_tmp = np.array(mag[i], dtype=float)
        if len(mag_tmp) > 1 and not np.isnan(mag_tmp).all():
            krt = np.nanvar(mag_tmp)
            kurtosis.append(krt)
        else:
            kurtosis.append(np.nan)
    return np.array(kurtosis)
```

```python
amplitude_study = amplituda(study_grupa, study_mag)
amplitude_control = amplituda(control_grupa, control_mag)
varijansa_study = varijansa(study_grupa, study_mag)
varijansa_control = varijansa(control_grupa, control_mag)

skewness_study = skewness(study_grupa, study_mag)
skewness_control = skewness(control_grupa, control_mag)
kurtosis_study = kurtosis(study_grupa, study_mag)
kurtosis_control = kurtosis(control_grupa, control_mag)

#bez nan
amplitude_st = amplitude_study[~np.isnan(amplitude_study)]
amplitude_ct = amplitude_control[~np.isnan(amplitude_control)]
varijanse_st = varijansa_study[~np.isnan(varijansa_study)]
varijanse_ct = varijansa_control[~np.isnan(varijansa_control)]

# 
skewness_st = skewness_study[~np.isnan(skewness_study)]
skewness_ct = skewness_control[~np.isnan(skewness_control)]
kurtosis_st = kurtosis_study[~np.isnan(kurtosis_study)]
kurtosis_ct = kurtosis_control[~np.isnan(kurtosis_control)]


#print('varijansa kontrolna',varijansa_control)
#print('varijansa study',varijansa_study)

print("-"*40)
#Studentov t-test
print("Studentov t-test")
t_stat, p_val = ttest_ind(amplitude_st, amplitude_ct, equal_var=False)
print('amplitude:','t-stat:',t_stat,"; p_vrednost: ",p_val)
t_stat, p_val = ttest_ind(varijanse_st, varijanse_ct, equal_var=False)
print('varijanse:','t-stat:',t_stat,"; p_vrednost: ",p_val)
t_stat, p_val = ttest_ind(skewness_st, skewness_ct, equal_var=False)
print('skewness:','t-stat:',t_stat,"; p_vrednost: ",p_val)
t_stat, p_val = ttest_ind(kurtosis_st, kurtosis_ct, equal_var=False)
print('kurtosis:','t-stat:',t_stat,"; p_vrednost: ",p_val)
print('-'*40)

# KS-test - Kolmogorov - Smirnov test
print('Kolmogorov-Smirnov test')
ks_stat, ks_p = ks_2samp(amplitude_st, amplitude_ct)
print('amplitude:','KS-stat:',ks_stat,";KS p_vrednost: ",ks_p)
ks_stat, ks_p = ks_2samp(varijanse_st, varijanse_ct)
print('varijanse:','KS-stat:',ks_stat,";KS p_vrednost: ",ks_p)
ks_stat, ks_p = ks_2samp(skewness_st, skewness_ct)
print('skewness:','KS-stat:',ks_stat,";KS p_vrednost: ",ks_p)
ks_stat, ks_p = ks_2samp(kurtosis_st, kurtosis_ct)
print('kurtosis:','KS-stat:',ks_stat,";KS p_vrednost: ",ks_p)
```

### kohen i klif test

```python
def cohens_d(group1, group2):

    n1, n2 = len(group1), len(group2)
    mean1, mean2 = np.mean(group1), np.mean(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)

    pooled_std = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    
    d = (mean1 - mean2) / pooled_std
    
    abs_d = abs(d)
    if abs_d < 0.2:
        interpretation = "negligible"
    elif abs_d < 0.5:
        interpretation = "small"
    elif abs_d < 0.8:
        interpretation = "medium"
    else:
        interpretation = "large"
    
    return d, interpretation

def cliffs_delta(group1, group2):
    
    n1, n2 = len(group1), len(group2)
    more = 0
    less = 0
    
    for x1 in group1:
        for x2 in group2:
            if x1 > x2:
                more += 1
            elif x1 < x2:
                less += 1

    delta = (more - less) / (n1 * n2)
    
    abs_delta = abs(delta)
    if abs_delta < 0.147:
        interpretation = "negligible"
    elif abs_delta < 0.33:
        interpretation = "small"
    elif abs_delta < 0.474:
        interpretation = "medium"
    else:
        interpretation = "large"
    
    return delta, interpretation
```

```python
def compare(study_group,control_group):
    study_group = np.array(study_group)
    control_group = np.array(control_group)

    t_stat, t_pval = ttest_ind(study_group, control_group)
    u_stat, u_pval = mannwhitneyu(study_group, control_group, alternative='two-sided')

    cohens, cohens_interp = cohens_d(study_group, control_group)
    cliffs, cliffs_interp = cliffs_delta(study_group, control_group)

    print(f"  Student's t-test: t = {t_stat:.4f}, p = {t_pval:.4f}")
    print(f"  Mann-Whitney U:   U = {u_stat:.4f}, p = {u_pval:.4f}")
    print(f"  Cohen's delta:    {cohens:+.4f}  ({cohens_interp})")
    print(f"  Cliff's Delta: {cliffs:+.4f}  ({cliffs_interp})")
```

### Pronalazak anomalija u svetlosnim krivama

```python
def detect_bursts(mag, ime, sigma_factor=2):
    
    mag = np.array(mag)
    median = np.median(mag)
    mad = np.median(np.abs(mag - median))
    sigma = 1.4826 * mad
    prag = median + sigma_factor * sigma

    bursts = mag > prag
    n_bursts = np.sum(bursts)

    if ime:
        print(f"{ime}:",f"Medijana: {median:.2f}",f"Prag ({sigma_factor}σ): {prag:.2f}",f"Broj burst tačaka: {n_bursts}")

    return bursts, prag, median, n_bursts
```

### fractional variability

```python
def compute_fvar(x, err):
    x = np.array(x)
    err = np.array(err)
    mean_x = np.mean(x)
    S2 = np.var(x, ddof=1)
    mean_err2 = np.mean(err**2)
    return np.sqrt((S2 - mean_err2) / mean_x**2) if S2 > mean_err2 else 0

def compare_fvar(study_mag,study_magerr,control_mag,control_magerr):
    fvar_study = []
    fvar_control = []
    for i in range(len(study_mag)):
        fvar1 = compute_fvar(study_mag[i],study_magerr[i])
        fvar_study.append(fvar1)
    for i in range(len(control_mag)):
        fvar2 = compute_fvar(control_mag[i],control_magerr[i])
        fvar_control.append(fvar2)
    
    #print(fvar_study)
    #print(fvar_control)
    return fvar_study,fvar_control
```

```python
fvar_study,fvar_control = compare_fvar(study_mag,study_magerr,control_mag,control_magerr)

plt.figure(figsize=(6, 5))
plt.boxplot([fvar_study, fvar_control], tick_labels=['Study', 'Control'],
           medianprops=dict(color='darkblue', linewidth=2))
plt.title('Fractional Varibility')
plt.ylabel('Fvar')
plt.grid(alpha=0.3)
plt.show()
```
